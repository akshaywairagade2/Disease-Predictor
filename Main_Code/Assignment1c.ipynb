Entropy

Here entropy of dataset is calculated as summation of 

-(number of times disease occurs/total number of diseases)*log2(number of times disease occurs/total number of diseases)

```python
import math
def entropy(dataframe):
  enpy=0
  dic=dict()
  rows=dataframe.shape[0]
  for i in dataframe.index:
    dise=df2.loc[i,'Disease']
    if dise in dic:
      dic[dise]+=1
    else:
      dic[dise]=1
  for i in dic:
    enpy+=(-dic[i]/rows)*math.log2(dic[i]/rows)
    # print(entropy)
  return enpy
```

Information Gain

Calculating entropy for positive value of disease in infpos and for 0 value of disease in infneg.

Applied the formula as teached in the class.

```python
def information_gain(disease,dataframe):
  posdic={}
  negdic={}
  tpo=0
  tne=0
  rows = dataframe.shape[0]
  for i in dataframe.index:
    if dataframe.loc[i][disease]==0:
      dise=dataframe.loc[i][0]
      if dise in negdic:
        negdic[dise]+=1
        tne+=1
      else:
        negdic[dise]=1
        tne+=1
    else:
      dise=dataframe.loc[i][0]
      if dise in posdic:
        posdic[dise]+=1
        tpo+=1
      else:
        posdic[dise]=1
        tpo+=1
  infpos=0
  infneg=0
  for i in posdic:
    infpos+=(-posdic[i]/tpo)*math.log2(posdic[i]/tpo)
  for i in negdic:
    infneg+=(-negdic[i]/tne)*(math.log2(negdic[i]/tne))
  try:
    final_gain=entropy(dataframe)-(tpo/(tpo+tne))*infpos-(tne/(tpo+tne))*infneg
  except:
    final_gain=0
  return final_gain

# information_gain('sinus_pressure',df2)
```

Calculated the information gain for the whole dataframe for all symptoms in dataframe.

```python
def information_gain_list(dataframe):
  lis=dataframe.columns
  lis=lis[1:]
  ansli=[]
  for i in lis:
    gain=information_gain(str(i),dataframe)
    ansli.append([gain,i])
  ansli.sort(reverse=True)
  return ansli
```

**The main function to create decision tree is here:-**

i)parent_child dictionary stores the parent node and its child node.

ii)decision_tree stores the nodes which are used in decision tree.

iii)Open List stores the nodes which are to be processed . This is done in queue.

**i)parent_child**

It stores the key as disease+'1' if it is obtained for disease value 0 and disease+'2' for disease value 1.

**ii)Decision_tree**

It stored the symptoms which are used in graph till now . And this symptoms are removed so that they do not come in decision_tree again.

**iii)Open List**

It stores the list of symptoms and their respective dataframes after splitting so that we do not have to split w.r.t. all its ancestors because that datframe will already be splitted.


```python
parent_child={}
decision_tree=[]
open_list=[]
y=0
# print("Hello World")
while(y==0):
  if decision_tree==[]:
    lis=information_gain_list(df2)
    decision_tree.append(lis[0][1])
    open_list.append([lis[0][1],df2])
    # print("Open List is ",open_list)
    print("Decision Tree Closed List is ",decision_tree)
    # print(parent_child)
    print(df2.shape[1])
    print("-------------------------------------------")
    print("\n\n\n")
  else:
    ele=open_list[0][0]
    dframe=open_list[0][1]
    open_list.pop(0)
    df_left=dframe[dframe[ele]<=0.5]
    df_right=dframe[dframe[ele]>=0.5]
    # print("Shape of left decision tree before ",df_left.shape[1])
    # print("Shape of right decision tree before ",df_right.shape[1])
    for i in range(len(decision_tree)):
      # ele=decision_tree[i]
      # print(decision_tree[i],type(decision_tree[i]))
      try:
        df_left = df_left.drop([decision_tree[i]], axis = 1)
        df_right = df_right.drop([decision_tree[i]],axis = 1)
      except:
        pass
    # print("Shape of left decision tree after ",df_left.shape[1])
    # print("Shape of right decision tree after ",df_right.shape[1])
    lis1=information_gain_list(df_left)
    lis2=information_gain_list(df_right)
    decision_tree.append(lis1[0][1])
    open_list.append([lis1[0][1],df_left])
    decision_tree.append(lis2[0][1])
    open_list.append([lis2[0][1],df_right])
    parent_child[ele+str(1)]=lis1[0][1]
    parent_child[ele+str(2)]=lis2[0][1]
    print(len(parent_child))
    print(parent_child)
    # print("Open List is ",open_list)
    print("Decision Tree Closed List is ",decision_tree)
    print("------------------------------------------")
    if (len(parent_child)==114):
      ansli=[]
      for i in parent_child:
        dise=i[:-1]
        true_false_indicator=i[-1]
        if true_false_indicator=='1':
          ansli.append([dise,parent_child[i],0])
        else:
          ansli.append([dise,parent_child[i],1])
      print(ansli)
      break
    print("\n\n\n")
    # print(parent_child)
```
